```python
import pandas as pd
```

```python
# we are specifying the response ID as the index while loading the data itself. We can use the set_index method later on as well as shown below
df = pd.read_csv('Data/survey_results_public.csv', index_col = "ResponseId")
schema_df = pd.read_csv('Data/survey_results_schema.csv')
```

```python
pd.set_option('display.max_columns', 79)
pd.set_option('display.max_rows', 79)
```

```python
df.head(5)
```

```python
# this is like a filter mask that will give you if every entry of that column meets (True) the criterion or false
df["YearsCode"] == "14"
```

```python
#When you apply this mask to a pdf. you get the filtered result.
df[df["YearsCode"] == "14"].head(2)
```

```python
#the same can be done using df.loc[filter]
```

```python
df.loc[df["YearsCode"] == "14"].head(2)
```

```python
# This filter is essentially for rows to be selected. Hence you can still pass a column name
df.loc[df["YearsCode"] == "14", ["CodingActivities", "YearsCode"]].head(2)
```

```python
#Multi filer
df.loc[(df["YearsCode"] == "14") & (df["CodingActivities"] == "Hobby"), ["CodingActivities", "YearsCode"]].head(2)

#i have no idea why and is not working instead of &
#df.loc[(df["YearsCode"] == "14") and (df["CodingActivities"] == "Hobby"), ["CodingActivities", "YearsCode"]].head(2)
```

```python
df.loc[(df["YearsCode"] == "14") | (df["CodingActivities"] == "Hobby"), ["CodingActivities", "YearsCode"]].head(5)
```

```python
# Another way is to define a filter seperately
filter = (df["YearsCode"] == "14") | (df["CodingActivities"] == "Hobby")
df[filter].head(5) 
```

```python
#we can also negate the filters
filter = (df["YearsCode"] == "14") | (df["CodingActivities"] == "Hobby")
df[~filter].head(5) 
```

```python
## EXAMPLES OF USING FILTERS
# I want salary greater than 215232. 
high_salary_filter = df['ConvertedCompYearly'] > 215232
df.loc[high_salary_filter, "ConvertedCompYearly"]
```

```python
# .ISIN() METHOD
countries = ["Germany", "India", "Israel", "Canada"]
#filter = df["Country"].isin(["Germany", "India", "Israel", "Canada"])
filter = df["Country"].isin(countries)
df[filter].head(2)
```

```python
#now i want to manipulate the data and get a subset of the column entry
#like if someomne knows "C#;JavaScript;SQL;TypeScript". I want it to be displayed in my SQL filter.

filter = df["LanguageHaveWorkedWith"].str.contains("SQL", na = False)
df[filter]
```
