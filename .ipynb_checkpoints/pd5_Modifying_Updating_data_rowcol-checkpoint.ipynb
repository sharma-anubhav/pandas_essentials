```python
# we are specifying the response ID as the index while loading the data itself. We can use the set_index method later on as well as shown below
df = pd.read_csv('Data/survey_results_public.csv', index_col = "ResponseId")
schema_df = pd.read_csv('Data/survey_results_schema.csv')
```

```python
people = {
    "fname" : ["Anubhav", "Saksham"],
    "lname" : ["Sharma", "Sharma"],
    "email" : ["anubhav@gmail.com", "saksham@gmail.com"]
}
people_df = pd.DataFrame(people)
people_df
```

```python
people_df.columns
```

```python
# Let us give new names to the columns
people_df.columns = ["First_name", "Last_name", "Email"]
people_df.columns
```

```python
# Perform operations on col names

#Col names UPPERCASe
people_df.columns = [x.upper() for x in people_df.columns]
people_df.columns

# replace _ with __
people_df.columns = [str(x).replace("_", "__") for x in people_df.columns]
#OR
people_df.columns = people_df.columns.str.replace("_", "__")

people_df.columns
```

```python
#usually we do not want to change everything about a column
```

```python
people_df.columns = [x.lower() for x in people_df.columns]
people_df.columns = people_df.columns.str.replace("__", "_")
```

```python
#changing  some columns only
#make sure to use inplace
people_df.rename(columns = {'first_name': 'First'}, inplace = True )
people_df
```

```python
#modifying data
people_df.loc[0, "First"] = "Anubhavs"
people_df.loc[0, ["First", "last_name"]] = ["Anubhavs", "sharmaa"]
people_df.loc[0] = ["Anubhavs", "sh", "anubhavs@gmail.com"]
people_df
```

```python
# for single value modifications you can use .at
people_df.at[0, "First"] = "Anubhavs"
people_df
```
