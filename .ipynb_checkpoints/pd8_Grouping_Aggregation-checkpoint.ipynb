```python
# but salaries vary by country So it would make more sense to see salary means by country
# hence the need for grouping

# groupby returns a group object which has multiple groups i.e data frames of those groups
# each object inside group object is a tuple and has 2 things (name and df)
country_groups = df.groupby(["Country"]) # we are passing a list as grouping can be nested
```

```python
country_groups.get_group("India")
# this is essentially similar to df[df["Country"] == "India"]
```

```python
# if we needed stats for a country basis we would do 
df[df["Country"] == "India"]["Employment"].value_counts()

# now using groupby object
country_groups["Employment"].value_counts()
    
```

```python
# we can also get one from the above grouping using .loc
country_groups["Employment"].value_counts().loc["India"]
```

```python
# General group stats can be seen directly by methods on group object
country_groups["ConvertedCompYearly"].mean()
```

```python
# Specific group can be called using get_group and then methods or just using .loc[]
country_groups["ConvertedCompYearly"].mean().loc["India"] 
country_groups.get_group("India")["ConvertedCompYearly"].mean()
```
