```python
people = {
    "fname" : ["Anubhav", "Saksham", "Suman"],
    "lname" : ["Sharma", "Sharma", "Kumar"],
    "email" : ["anubhav@gmail.com", "saksham@gmail.com", "drsuman@gmail.com"],
    "age" : [1,2,3]
}
people_df = pd.DataFrame(people)
people_df
```

```python
# Adding new column and concat of existing 
people_df["full_name"] = people_df["fname"] + " " + people_df["lname"]
people_df
```

```python
# Another good example is adding new column using a math operation and apply method and other operations from earlier
people_df["new_Age"] = people_df["age"] + 10
people_df["new_Age3"] = people_df["age"].apply(lambda x: x+20)
people_df
```

```python
# .insert 
# adds new col at a given position
people_df.insert(1, "new_Age4" , [31,32,33])
```

```python
#DROP col
people_df.drop(columns = ["new_Age3", "new_Age", "new_Age4"], inplace = True)
people_df
```

```python
# Split and Expand
people_df[["new_firstName", "new_lastName"]] = people_df["full_name"].str.split(" ", expand = True)
people_df
```

```python
# ADD ROWS and APPENDING DF's

# things to keep in mind. adding row needs a index. but if you dont want to give one, set ignore_index = false and then add.
# the values you do not give in a dict are set to nan

#df.append({"fname": "Anupama", "name": "Sharma"})
# ******** .APPEND METHOD WAS DEPRICATED FROM LATEST PANDAS. WE DONT USE IT NOW

```

```python
# .loc
# data that you dont give is set to NaN
people_df.loc[3] = {"fname" : "Anupama"}
people_df
```

```python
# making extra data to concat
people2 = {
    "fname" : ["Anubhav3", "Saksham3", "Suman3"],
    "email" : ["anubhav3@gmail.com", "saksham3@gmail.com", "drsuman3@gmail.com"],
}
people_df2 = pd.DataFrame(people2)
```

```python
# .CONCAT
# the columns that are not there are set to NaN
people_df = pd.concat([people_df, people_df2])
people_df
```

```python
# Duplicate indexes seen
people_df.loc[0]
```

```python
# now it will continue the index series 
people_df = pd.concat([people_df, people_df2], ignore_index = True)
people_df
```

```python
# Delete Rows
people_df.drop(index = 4, inplace = True)
```

```python
# deleting ROWS with FILTERS
filter = people_df["lname"] == "Sharma"
indexes = people_df[filter].index

people_df.drop(index = indexes, inplace = True)
people_df
```
